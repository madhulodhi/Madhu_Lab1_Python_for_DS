```python
import mysql.connector

# MySQL connection details
host = "localhost"
user = "root"  # Change this to your MySQL username
password = "root"  # Change this to your MySQL password
database = "e_commerce"

# Connect to MySQL
conn = mysql.connector.connect(host=host, user=user, password=password, database=database)

if conn.is_connected():
    print("Connected to MySQL database")
else:
    print("Failed to connect to MySQL database")

# Create a cursor object for each table
cursor_supplier = conn.cursor()
cursor_customer = conn.cursor()
cursor_category = conn.cursor()
cursor_product = conn.cursor()
cursor_product_details = conn.cursor()
cursor_order = conn.cursor()
cursor_rating = conn.cursor()

# Create Supplier Table
create_supplier_table = """
CREATE TABLE IF NOT EXISTS supplier (
    SUPP_ID INT PRIMARY KEY,
    SUPP_NAME VARCHAR(50),
    SUPP_CITY VARCHAR(50),
    SUPP_PHONE VARCHAR(10)
)
"""
cursor_supplier.execute(create_supplier_table)
print("Supplier table created successfully")

# Create Customer Table
create_customer_table = """
CREATE TABLE IF NOT EXISTS customer (
    CUS_ID INT NOT NULL PRIMARY KEY,
    CUS_NAME VARCHAR(20) DEFAULT NULL,
    CUS_PHONE VARCHAR(10),
    CUS_CITY VARCHAR(30),
    CUS_GENDER CHAR
)
"""
cursor_customer.execute(create_customer_table)
print("Customer table created successfully")

# Create Category Table
create_category_table = """
CREATE TABLE IF NOT EXISTS category (
    CAT_ID INT NOT NULL PRIMARY KEY,
    CAT_NAME VARCHAR(20) DEFAULT NULL
)
"""
cursor_category.execute(create_category_table)
print("Category table created successfully")

# Create Product Table
create_product_table = """
CREATE TABLE IF NOT EXISTS product (
    PRO_ID INT NOT NULL PRIMARY KEY,
    PRO_NAME VARCHAR(20) DEFAULT NULL,
    PRO_DESC VARCHAR(60) DEFAULT NULL,
    CAT_ID INT NOT NULL,
    FOREIGN KEY (CAT_ID) REFERENCES category(CAT_ID)
)
"""
cursor_product.execute(create_product_table)
print("Product table created successfully")

# Create Product_Details Table
create_product_details_table = """
CREATE TABLE IF NOT EXISTS product_details (
    PROD_ID INT NOT NULL,
    PRO_ID INT NOT NULL,
    SUPP_ID INT NOT NULL,
    PROD_PRICE INT NOT NULL,
    PRIMARY KEY (PROD_ID),
    FOREIGN KEY (PRO_ID) REFERENCES product(PRO_ID),
    FOREIGN KEY (SUPP_ID) REFERENCES supplier(SUPP_ID)
)
"""
cursor_product_details.execute(create_product_details_table)
print("Product_Details table created successfully")

# Create Order Table
create_order_table = """
CREATE TABLE IF NOT EXISTS orders (
    ORD_ID INT NOT NULL PRIMARY KEY,
    ORD_AMOUNT INT NOT NULL,
    ORD_DATE DATE,
    CUS_ID INT NOT NULL,
    PROD_ID INT NOT NULL,
    FOREIGN KEY (CUS_ID) REFERENCES customer(CUS_ID),
    FOREIGN KEY (PROD_ID) REFERENCES product_details(PROD_ID)
)
"""
cursor_order.execute(create_order_table)
print("Order table created successfully")

# Create Rating Table
create_rating_table = """
CREATE TABLE IF NOT EXISTS rating (
    RAT_ID INT NOT NULL PRIMARY KEY,
    CUS_ID INT NOT NULL,
    SUPP_ID INT NOT NULL,
    RAT_RATSTARS INT NOT NULL,
    FOREIGN KEY (SUPP_ID) REFERENCES supplier(SUPP_ID),
    FOREIGN KEY (CUS_ID) REFERENCES customer(CUS_ID)
)
"""
cursor_rating.execute(create_rating_table)
print("Rating table created successfully")

# Close all cursors and connection
cursor_supplier.close()
cursor_customer.close()
cursor_category.close()
cursor_product.close()
cursor_product_details.close()
cursor_order.close()
cursor_rating.close()
conn.close()
```

```python
import mysql.connector

# MySQL connection details
host = "localhost"
user = "root"  
password = "root"  
database = "e_commerce"

# Connect to MySQL
conn = mysql.connector.connect(host=host, user=user, password=password, database=database)

if conn.is_connected():
    print("Connected to MySQL database")
else:
    print("Failed to connect to MySQL database")

try:
    # SQL query
    sql_query = """
    SELECT customer.CUS_GENDER, COUNT(DISTINCT orders.CUS_ID) AS num_customers
    FROM orders
    JOIN customer ON orders.CUS_ID = customer.CUS_ID
    WHERE ORD_AMOUNT >= 3000
    GROUP BY customer.CUS_GENDER;
    """

    # Execute the SQL query
    cursor = conn.cursor()
    cursor.execute(sql_query)

    # Fetch all the results
    results = cursor.fetchall()

    # Display the results
    print("Number of customers grouped by gender who have placed any order of amount greater than or equal to Rs. 3000:")
    print("Gender   |   Number of Customers")
    print("-----------------------------")
    for row in results:
        print(f"{row[0]}       |   {row[1]}")

except mysql.connector.Error as error:
    print("Error executing SQL query:", error)

finally:
    # Close the cursor and connection
    if 'cursor' in locals():
        cursor.close()
    conn.close()
```

```python
### Q5) Display the Supplier details who can supply more than 
##one product.
import mysql.connector

# MySQL connection details
host = "localhost"
user = "root"  
password = "root"  
database = "e_commerce"

# Connect to MySQL
conn = mysql.connector.connect(host=host, user=user, password=password, database=database)

if conn.is_connected():
    print("Connected to MySQL database")
else:
    print("Failed to connect to MySQL database")

try:
    # SQL query
    sql_query = """
    SELECT supplier.*
    FROM supplier
    JOIN product_details ON supplier.SUPP_ID = product_details.SUPP_ID
    GROUP BY supplier.SUPP_ID
    HAVING COUNT(product_details.PRO_ID) > 1;
    """

    # Execute the SQL query
    cursor = conn.cursor()
    cursor.execute(sql_query)

    # Fetch all the results
    results = cursor.fetchall()

    # Display the results
    print("Supplier details who can supply more than one product:")
    print("---------------------------------------------------------------------------")
    print("Supplier ID | Supplier Name | Supplier City | Supplier Phone")
    print("---------------------------------------------------------------------------")
    for row in results:
        print(f"{row[0]}          |   {row[1]}      |   {row[2]}   |   {row[3]}")

except mysql.connector.Error as error:
    print("Error executing SQL query:", error)

finally:
    # Close the cursor and connection
    if 'cursor' in locals():
        cursor.close()
    conn.close()

```

```python
import mysql.connector

# MySQL connection details
host = "localhost"
user = "root"  
password = "root"  
database = "e_commerce"

# Connect to MySQL
conn = mysql.connector.connect(host=host, user=user, password=password, database=database)

if conn.is_connected():
    print("Connected to MySQL database")
else:
    print("Failed to connect to MySQL database")

try:
    # SQL query
    sql_query = """
    SELECT category.CAT_NAME
    FROM orders
    JOIN product_details ON orders.PROD_ID = product_details.PROD_ID
    JOIN product ON product_details.PRO_ID = product.PRO_ID
    JOIN category ON product.CAT_ID = category.CAT_ID
    WHERE orders.ORD_AMOUNT = (SELECT MIN(ORD_AMOUNT) FROM orders);
    """

    # Execute the SQL query
    cursor = conn.cursor()
    cursor.execute(sql_query)

    # Fetch all the results
    results = cursor.fetchall()

    # Display the results
    print("Category of the product with the minimum order amount:")
    for row in results:
        print(row[0])

except mysql.connector.Error as error:
    print("Error executing SQL query:", error)

finally:
    # Close the cursor and connection
    if 'cursor' in locals():
        cursor.close()
    conn.close()
```

```python
### Q8) Print the top 3 supplier name and id and rating on the basis of their rating along with the customer name who has given 
##the rating.

import mysql.connector

# MySQL connection details
host = "localhost"
user = "root"  # Change this to your MySQL username
password = "root"  # Change this to your MySQL password
database = "e_commerce"

# Connect to MySQL
conn = mysql.connector.connect(host=host, user=user, password=password, database=database)

if conn.is_connected():
    print("Connected to MySQL database")
else:
    print("Failed to connect to MySQL database")

try:
    # SQL query
    sql_query = """
    SELECT supplier.SUPP_ID, supplier.SUPP_NAME, rating.RAT_RATSTARS, 
    customer.CUS_NAME
    FROM rating
    JOIN supplier ON rating.SUPP_ID = supplier.SUPP_ID
    JOIN customer ON rating.CUS_ID = customer.CUS_ID
    ORDER BY rating.RAT_RATSTARS DESC
    LIMIT 3;
    """

    # Execute the SQL query
    cursor = conn.cursor()
    cursor.execute(sql_query)

    # Fetch all the results
    results = cursor.fetchall()

    # Display the results
    print("Top 3 suppliers along with their ratings and the names of the customers who have given the ratings:")
    print("-----------------------------------------------------------------------------------------------------")
    print("Supplier ID | Supplier Name | Rating | Customer Name")
    print("-----------------------------------------------------------------------------------------------------")
    for row in results:
        print(f"{row[0]}          |   {row[1]}      |   {row[2]}    |   {row[3]}")

except mysql.connector.Error as error:
    print("Error executing SQL query:", error)

finally:
    # Close the cursor and connection
    if 'cursor' in locals():
        cursor.close()
    conn.close()
```
